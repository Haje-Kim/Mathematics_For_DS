# **Linear Algebra and Explain Geometry**

## **The Geometric meaning of the Vector**

* In $n$-dimension space, a $n$-dimension vector $a$ is:
    * A **point** described by the values of vector $a$
    * An **arrow** that connects starting point with vector $a$ point

* Keeping its length and direction stable, we can parallelly translate a vector 

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
plt.rc('font', size=18)
gray = {'facecolor' : 'gray'}
black = {'facecolor' : 'black'}
red = {'facecolor' : 'red'}
green = {'facecolor' : 'green'}
blue = {'facecolor' : 'blue'}
```

```python
a = np.array([1, 2])
plt.plot(0, 0, 'kP', ms=20)
plt.plot(a[0], a[1], 'ro', ms=20)
plt.annotate('', xy=[-0.6, 1.6], xytext=(0.2, 0.7), arrowprops=gray)
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=black)
plt.annotate('', xy=a + [-1, 1], xytext=(-1, 1), arrowprops=black)
plt.text(0.35, 1.15, "$a$")
plt.text(1.15, 2.25, "$(1,2)$")
plt.text(-0.7, 2.1, "$a$")
plt.text(-0.9, 0.6, "Parallel \n Movement")
plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-2.4, 3.4)
plt.ylim(-0.8, 3.4)
plt.show()
```

## **Vector Length**

* $||a|| = \sqrt{a^Ta} = \sqrt{a_1^2+\cdot \cdot \cdot + a_N^2}$ = the length of vector $a$

```python
a = np.array([1,2])
np.linalg.norm(a)
```

## **Vector mutiplied by Scalar**

```python
a = np.array([1, 2])
b = 2 * a
c = -a
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=red)
plt.text(0.8, 3.1, "$2a$")
plt.text(2.2, 3.8, "$(2, 4)$")
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)
plt.text(0.1, 1.3, "$a$")
plt.text(1.1, 1.4, "$(1, 2)$")
plt.plot(c[0], c[1], 'ro', ms=10)
plt.annotate('', xy=c, xytext=(0, 0), arrowprops=blue)
plt.text(-1.3, -0.8, "$-a$")
plt.text(-3, -2.5, "$(-1, -2)$")
plt.plot(0, 0, 'kP', ms=20)
plt.xticks(np.arange(-5, 6))
plt.yticks(np.arange(-5, 6))
plt.xlim(-4.4, 5.4)
plt.ylim(-3.2, 5.2)
plt.show()
```

## **Unit Vector**

* The vector with length = 1
* e.g. $a=\begin{bmatrix} 1 \\ 0\end{bmatrix}$, $b=\begin{bmatrix} 0 \\ 1\end{bmatrix}$, $c=\begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}}\end{bmatrix}$

* For a random vector $x$ that is not a zero vector, 
* The following vector is a unit vector with the same direction as vector $x$
    * $\frac{x}{||x||}$

```python
a = np.array([1,0])
b = np.array([0,1])
c = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
np.linalg.norm(a), np.linalg.norm(b), np.linalg.norm(c)
```

## **The sum of Vector**

```python
a = np.array([1, 2])
b = np.array([2, 1])
c = a + b
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=c, xytext=(0, 0), arrowprops=black)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.plot(c[0], c[1], 'ro', ms=10)
plt.plot([a[0], c[0]], [a[1], c[1]], 'k--')
plt.plot([b[0], c[0]], [b[1], c[1]], 'k--')
plt.text(0.35, 1.15, "$a$")
plt.text(1.15, 0.25, "$b$")
plt.text(1.25, 1.45, "$c$")
plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.show()
```

## **The Linear Combination of Vector**

* A linear combination is the sum of multiple vector multiplied by scalar 

* $c_1x_1+c_2x_2+\cdot \cdot \cdot +c_nx_n$
* the $c_1, \cdot \cdot \cdot, c_N$ is the scalar coefficient 

```python
x1 = np.array([1, 2])
x2 = np.array([2, 1])
x3 = 0.5 * x1 + x2
plt.annotate('', xy=0.5*x1, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=x2, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=x3, xytext=(0, 0), arrowprops=black)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(x1[0], x1[1], 'ro', ms=10)
plt.plot(x2[0], x2[1], 'ro', ms=10)
plt.plot(x3[0], x3[1], 'ro', ms=10)
plt.plot([x1[0], 0], [x1[1], 0], 'k--')
plt.text(0.6, 2.0, "$x_1$")
plt.text(-0.5, 0.5, "$0.5x_1$")
plt.text(1.15, 0.25, "$x_2$")
plt.text(2.5, 1.6, "$0.5x_1 + x_2$")
plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.show()
```

## **Euclidean Distance**

* Euclidean Distance is the distance between the points of two vectors 
* The euclidean distance can be derived by the length difference of vectors 

* $||a-b|| = \sqrt{\Sigma{(a_i-b_i)^2}}=\sqrt{||a||^2+||b||^2-2a^Tb}$
* $||a-b||^2 = ||a||+||b||-2a^Tb$

## **The inner product of vector, and trigonometric functions**

* $a^Tb = ||a||||b||cos\theta$

* $sin\theta=\frac{a}{h}$

* $sin 0\circ = 0$, $sin 90\circ = 1$

* $cos\theta=\frac{b}{h}$
* $cos 0\circ = 1$, $cos 90 \circ = 0$

```python
plt.plot([0, 1], [0, 2], 'k-', lw=3)
plt.plot([0, 1], [0, 0], 'k-', lw=3)
plt.plot([1, 1], [0, 2], 'k-', lw=3)
plt.text(0.05, 1, "Hypotenuse, h")
plt.text(0.35, -0.2, "Opposite, b")
plt.text(1.05, 1, "Adjacent, a")
plt.text(0.12, 0.06, r"$\theta$")
plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.1, 2.1)
plt.ylim(-0.5, 2.3)
plt.show()
```

```python
x = np.linspace(0, np.pi/2, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, 'r--', lw=3,  label=r"$\sin\theta$")
plt.plot(x, y2, 'b-', lw=3, label=r"$\cos\theta$")
plt.legend()
plt.xticks([0, np.pi/4, np.pi/2], [r'$0^{\circ}$', r'$45^{\circ}$', r'$90^{\circ}$'])
plt.xlabel(r"$\theta$")
plt.title(r"The graph of $\sin\theta$ and $\cos\theta$")
plt.show()
```

## **Orthogonal**

* Orthogonal $a\perp b$ is when the angle from two vectors $a$ and $b$ equals 90 degrees

* Because $cos$ 90 $\circ$ = 0, the inner product of $a \perp b$ = 0
* $a^Tb = b^Ta = 0 \leftrightarrow a\perp b$

```python
a = np.array([1, 1])
b = np.array([-1, 1])

a.T @ b
```

## **Orthonormal**

* When $N$ number of unit vector $v_1,v_2, \cdot \cdot \cdot v_N$ are orthogonal to each other, it is called **orthonormal**

* $||v_i||=1 \leftrightarrow v_i^Tv_i = 1$
* $v_i^Tv_j=0 (i\ne j)$

## **Cosine Similarity**

* Cosine Similarity is the cosine value of the inner angle of two vectors 

* Since cosine is largest when $\theta = 0$, 
* When two vectors are pointing the same direction, the cosine similarity is largest at $1$

* Cosine Similarity = $cos \theta = \frac{x^Ty}{||x||||y||}$

* Cosine Distance = 1 - Cosine Similarity = 1 - $\frac{x^Ty}{||x||||y||}$

```python
a = np.array([4,5,2,2])
b = np.array([4,0,2,0])
c = np.array([2,2,0,1])

ed_ab = np.linalg.norm(a-b)
ed_ac = np.linalg.norm(a-c)
ed_bc = np.linalg.norm(b-c)

ed_ab, ed_ac, ed_bc
```

```python
cd_ab = 1 - (a.T @ b) / (np.linalg.norm(a) * np.linalg.norm(b))
cd_ac = 1 - (a.T @ c) / (np.linalg.norm(a) * np.linalg.norm(c))
cd_bc = 1 - (b.T @ c) / (np.linalg.norm(b) * np.linalg.norm(c))

cd_ab, cd_ac, cd_bc
```
